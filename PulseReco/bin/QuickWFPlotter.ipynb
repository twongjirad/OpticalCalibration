```python
%matplotlib inline
#%pylab
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
from root_numpy import root2array, root2rec, tree2rec, array2root
import ROOT
```

```python
#fname='../../../../temp_data/20150802_flasher/wf_test2.root'
fname='/Users/twongjirad/working/uboone/data/FlasherData_080715/wf_run003.root'
#fname='wf_test123_pt_trig.root'
numpy_rec_array = root2array(fname,'raw_wf_tree')

wf_df=pd.DataFrame(numpy_rec_array)
wf_df
```

```python
target_ch = range(0,1)
logic_id = (6,39)

for key,ev_df in wf_df.groupby('event'):
    if key>20:
        continue
    print 'event:',key
    # plot logic ch wf
    logic_wf_sr = ev_df.query('slot==%d and ch==%d' % logic_id)['wf']
    logic_wf = logic_wf_sr[logic_wf_sr.first_valid_index()]
    print type(logic_wf)

    for ch in target_ch:
        wf1 = ev_df.query('slot==5 and ch==%d' % ch)['wf']
        wf2 = ev_df.query('slot==6 and ch==%d' % ch)['wf']
        print wf1.first_valid_index(), wf2.first_valid_index()
        wf1 = wf1[wf1.first_valid_index()]
        wf2 = wf2[wf2.first_valid_index()]
        
        ymax = 2150
        #ymax = logic_wf.max()
        #if wf1.max() > ymax: ymax = wf1.max()
        #if wf2.max() > ymax: ymax = wf2.max()

        ymin = 2030
        #ymin = logic_wf.min()
        if wf1.min() < ymin: ymin = wf1.min()
        if wf2.min() < ymin: ymin = wf2.min()
            
        #ymax *= 1.1

        plt.figure(figsize=(14,6))
        plt.plot(np.arange(0,len(logic_wf),1),
                 logic_wf+20,
                 color='black')
        plt.plot(np.arange(0,len(wf1),1),wf1,color='blue')
        plt.plot(np.arange(0,len(wf2),1),wf2,color='red')

        plt.ylim(ymin,ymax)
        plt.tick_params(labelsize=20)
        plt.xlabel('sample',fontsize=20)
        plt.ylabel('ADC',fontsize=20)
        plt.grid()
        plt.title('Ch. %d Logic (black) HG (blue) LG (red)' % ch,fontsize=20)
        x1,x2,y1,y2 = plt.axis()
        plt.axis( (x1,x2,ymin,ymax))
        plt.show()

    #break
    
```
